```python
import matplotlib.pyplot as plt
import numpy as np

k = 8

x = []
y = []

for i in range(1001):
    for j in range(1001):

        r1 = (abs(-5 + j * 0.01) ** 2 + abs(-5 + i * 0.01 + 1) ** 2) ** 0.5
        r2 = (abs(-5 + j * 0.01) ** 2 + abs(-5 + i * 0.01 - 1) ** 2) ** 0.5

        
        if r1 != 0 and r2 != 0 and abs(1/r1 + 1/r2 * 10 - k) < 0.1:
            x.append(-5 + i * 0.01)
            y.append(-5 + j * 0.01)
```

```python
plt.scatter(x, y, s = 0.1)
plt.ylim(-5, 5)
plt.xlim(-5, 5)
```
